```python
import os, getpass, pandas as pd
from sqlalchemy import create_engine, text

SHIPS_DB_URL = "postgresql://neondb_owner:npg_CyJZE36GaBQo@ep-old-king-adwymhup-pooler.c-2.us-east-1.aws.neon.tech/neondb?sslmode=require&channel_binding=require"
engine = create_engine(SHIPS_DB_URL, pool_pre_ping=True, future=True)
```

```python
create_classes = """
    CREATE TABLE IF NOT EXISTS classes(
      class VARCHAR(25),
      type VARCHAR(25),
      country VARCHAR(25),
      numGuns INT,
      bore INT,
      displacement INT
    )
"""

create_battles = """
    CREATE TABLE IF NOT EXISTS battles(
      name VARCHAR(25),
      date DATE
    )
"""

create_outcomes = """
    CREATE TABLE IF NOT EXISTS outcomes(
      ship VARCHAR(25),
      battle VARCHAR(25),
      result VARCHAR(25)
    )
"""

create_ships = """
    CREATE TABLE IF NOT EXISTS ships(
      name VARCHAR(25),
      class VARCHAR(25),
      launched INT
    )
"""
```

```python
with engine.connect() as conn:
    conn.execute(text(create_classes))
    conn.execute(text(create_battles))
    conn.execute(text(create_outcomes))
    conn.execute(text(create_ships))
    conn.commit()
    print("Tables created")
```

```python
check_tables = """
    SELECT * FROM information_schema.tables
    WHERE table_name IN ('classes', 'battles', 'outcomes', 'ships');
"""

with engine.connect() as conn:
    check_tables = pd.read_sql_query(text(check_tables), conn)

display(check_tables)
```
